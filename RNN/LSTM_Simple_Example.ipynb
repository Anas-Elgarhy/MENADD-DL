```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
x_train = [[i/100 for i in range(20)]]
print('x_train',x_train)
x_train = np.array(x_train, dtype=float)
x_train = x_train.reshape((1, 1, 20)) 

y_train = [[i/100 for i in range(1,21)]]
print('y_train',y_train)
y_train = np.array(y_train, dtype=float)
y_train = y_train.reshape((1, 1, 20))

x_test=[i/100 for i in range(20,40)]
print('x_test',x_test)
x_test=np.array(x_test).reshape((1,1,20));

y_test=[i/100 for i in range(21,41)]
print('y_test',y_test)
y_test=np.array(y_test).reshape(1,1,20)
```

```python
model = Sequential()  
model.add(LSTM(150, input_shape=(1, 20),return_sequences=True))#Whether to return the last output in the output sequence, or the full sequence.
model.add(Dense(20))
model.compile(loss='mean_absolute_error', optimizer='adam',metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1600, batch_size=1, verbose=2)
```

```python
predict = model.predict(x_train)
predict_test = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(20), predict, c='red')
plt.scatter(range(20), y_train, c='green')
```

```python
plt.scatter(range(20), predict_test, c='blue')
plt.scatter(range(20), y_test, c='black')
```
